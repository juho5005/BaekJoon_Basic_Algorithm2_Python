### 인접 행렬로 해결한 방식

```python
# DFS와 BFS 
# 단, 방문할 수 있는 정점이 여러 개인 경우는 정점 번호가 작은 것부터 방문한다.
# 더 이상한 방문할 수 있는 점이 없는 경우 종료한다.
# 정점 번호는 1번 ~ N번까지이다.

from collections import deque
import sys 
# 정점, 간선, 탐색을 시작할 정점의 번호 
n, m, v = tuple(map(int, sys.stdin.readline().split()))
q = deque()

linked_vertex = [
    tuple(map(int, sys.stdin.readline().split()))
    for _ in range(m)
]

graph = [
    [0 for _ in range(n+1)]
    for _ in range(n+1)
]

# 간선(양방향)에 연결돼있는 graph를 1로 초기화 시켜준다.
for start, end in linked_vertex :
    graph[start][end] = 1
    graph[end][start] = 1

dfs_visited = [
    False 
    for _ in range(n+1)
]

bfs_visited = [
    False 
    for _ in range(n+1)
]

def dfs(curr_v) :
    for next_v in range(1, n+1) :
        if graph[curr_v][next_v] and not dfs_visited[next_v] :
            dfs_seq.append(next_v)
            dfs_visited[next_v] = True 
            dfs(next_v)


def bfs() :
    while q : 
        curr_v = q.popleft()
        for next_v in range(1, n+1) :
            if graph[curr_v][next_v] and not bfs_visited[next_v] :
                bfs_seq.append(next_v)
                bfs_visited[next_v] = True
                q.append(next_v)


root_vertex = v
dfs_visited[root_vertex] = True 
bfs_visited[root_vertex] = True

dfs_seq = [root_vertex]
bfs_seq = [root_vertex]

dfs(root_vertex)
q.append(root_vertex)
bfs()

print(*dfs_seq)
print(*bfs_seq)
```

### 인접 리스트로 해결한 방식 (인접행렬에 비해 메모리와 시간효율이 좋다)

```python
# DFS와 BFS 
# 단, 방문할 수 있는 정점이 여러 개인 경우는 정점 번호가 작은 것부터 방문한다.
# 더 이상한 방문할 수 있는 점이 없는 경우 종료한다.
# 정점 번호는 1번 ~ N번까지이다.

from collections import deque
import sys 
# 정점, 간선, 탐색을 시작할 정점의 번호 
n, m, v = tuple(map(int, sys.stdin.readline().split()))
q = deque()

linked_vertex = [
    tuple(map(int, sys.stdin.readline().split()))
    for _ in range(m)
]

graph = [
    []
    for _ in range(n+1)
]

# 간선(양방향)에 연결돼있는 graph를 1로 초기화 시켜준다.
for start, end in linked_vertex :
    graph[start].append(end)
    graph[end].append(start)

for g in graph : 
    g.sort() # 각 인접 리스트마다 오름차순으로 정렬을 해준다.

dfs_visited = [
    False 
    for _ in range(n+1)
]

bfs_visited = [
    False 
    for _ in range(n+1)
]

def dfs(curr_v) :
    for next_v in graph[curr_v] :
        if not dfs_visited[next_v] :
            dfs_seq.append(next_v)
            dfs_visited[next_v] = True 
            dfs(next_v)


def bfs() :
    while q : 
        curr_v = q.popleft()
        for next_v in graph[curr_v] :
            if not bfs_visited[next_v] :
                bfs_seq.append(next_v)
                bfs_visited[next_v] = True
                q.append(next_v)

root_vertex = v
dfs_visited[root_vertex] = True 
bfs_visited[root_vertex] = True

dfs_seq = [root_vertex]
bfs_seq = [root_vertex]

dfs(root_vertex)
q.append(root_vertex)
bfs()

print(*dfs_seq)
print(*bfs_seq)
```
